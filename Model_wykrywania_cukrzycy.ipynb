# **1. WSTĘP**
Na całym świecie rośnie liczba osób chorujących na cukrzyce. W Polsce w 2017 roku odnotowano ich ponad 2,5 miliona [1]. Jest to poważna, przeklekła choroba cywilizacyjna. Dziewięć na dziesięć osób chorych leczy się na cukrzyce typu drugiego, której przyczynami moga być mała aktywność fizyczna, nieodpowiednia dieta lub otyłość. Cukrzyca występuje, kiedy trzustka nie wytwarza wystarczającej ilości insuliny albo jeśli nasz organizm nie potrafi dobrze korzystać z wytwarzanej insuliny. [2]

Symptomy [3] z jakimi może się zmagać osoba z nierozpoznaną cukrzycą to m.in.:

*   uczucie wzmożonego pragnienia;
*   utrata masy ciała;
*   wielomocz;
*   przewlekłe zmęczenie;
*   podatność na częste infekcje grzybicze;
*   niewyraźne widzenie.

## **Notka historyczna**
Symptomy cukrzycy były opisywane już ponad 3000 tysięcy lat temu w starożytnym Egipcie, literaturze medycznej starożytnych Indii i Chin, również w pracach pisanych przez starożytnych arabskich i greckich lekarzy. Słowo "mellitus" (z łac. słodki jak miód) zostało dodane do łacińskiej nazwy po ponownym odkryciu słodkiego smaku moczu i krwi u pacjentów. Dopiero ponad 100 lat pózniej potwierdzono, że przyczyną tego słodkiego smaku jest nadmiar cukru. Przełomowym dla diabetyków wydarzeniem było wyizolowanie insuliny przez dwóch kanadyjskich naukowców w 1921 roku. Pozwoliło to ułatwić i uratować życia milionów ludzi w przyszłości. [4]

## **Cel i motywacja**
Zarówno skrajnie wysokie, jak i skrajnie niskie stężenia glukozy we krwi wiążą się z ciężkimi objawami patologicznymi. W ten sposób regulacja i kontrola poziomu glukozy we krwi są zasadniczą funkcją organizmu. W latach 60. XX wieku naukowcy i lekarze z Minnesoty (dr Rosevear i Molnar z *Mayo Clinic* oraz Ackerman i Gatewood z *University of Minnesota*) pracowali nad skonstruowaniem modelu matematycznego, który dokładnie opisywałby układ regulacji poziomu glukozy we krwi podczas badania krzywej cukrowej. Opracowali prosty model, w którym dwa parametry dają kryteria odróżniania stanu przedcukrzycowego od cukrzycy typu drugiego, co znacznie ułatwia prawidłową diagnozę, a później leczenie. [5]


# **2. TEORIA**
**Jak wygląda proces wykrywania cukrzycy typu 2 u pacjenta?** [6]

Należy wykonać **badanie krzywej cukrowej**, w którym pobierany jest poziom glukozy z krwi żylnej pacjenta na czczo oraz po 2h od spożycia 75 gram dawki glukozy (cukru). **Matematyczna analiza** zmiany kluczowych hormonów wpływających na regulację poziomu glukozy we krwi określa kryteria potrzebne do prawidłowej diagnozy.

**Jakie hormony wpływają na nasz poziom cukru?**

Insulina jest kluczowym hormonem wpływającym na poziom glukozy - obniża go. Glukagon, adrenalina, kortyzol, tyroksyna i hormon wzrostu go podwyższają. Przy prawidłowej pracy organizmu normoglikemia jest cały czas zachowana, przy jej zaburzeniach lekarz stawia diagnozę - stan przedcukrzycowy bądź cukrzyca typu 2.

**Jak wygląda podstawowy model układu regulacji glukozy we krwi?**

Jest opisany analitycznie za pomocą dwóch równań:

\begin{align}
\frac{\operatorname{d}\!{G}}{\operatorname{d}\!t} &= F_1(G,H)+J(t), \\
\frac{\operatorname{d}\!{H}}{\operatorname{d}\!t} &= F_2(G,H),
\end{align}
gdzie:
- $G$ to poziom glukozy we krwi ($G_0$ to optymalna wartość),
- $H$ to stężenie hormonów netto ($H_0$, to optymalna wartość),
- $J(t)$ to funkcja określająca szybkość, z jaką zwiększa się stężenie glukozy we krwi.

# **3. ROZWIĄZANIE ANALITYCZNE**
### **Początkowe założenia**
Pacjent na czczo ma prawidłowe wartości poziomu glukozy oraz stężenia hormonów netto czyli $G_0$ i $H_0$, zatem $F_1(G_0,H_0)=0$ oraz $F_2(G_0,H_0)=0$.

### **Szukanie odchylenia**
W badaniu interesuje nas odchylenie od $G$ i $H$, co oznaczymy przez $g=G-G_0$ i $h=H-H_0$. Wtedy możemy zapisać
\begin{align}
\frac{\operatorname{d}\!{g}}{\operatorname{d}\!t} &= F_1(G_0+g,H_0+h)+J(t), \\
\frac{\operatorname{d}\!{h}}{\operatorname{d}\!t} &= F_2(G_0+g,H_0+h).
\end{align}

Przez badanie krzywej cukrowej możemy zdiagnozować tylko stan przedcukrzycowy lub cukrzycę typu 2, dlatego możemy założyć, że $G$ i $H$ mają małe odchylenie od $G_0$ i $H_0$. Zatem przekształcamy powyższe równania następująco
\begin{align}
\frac{\operatorname{d}\!{g}}{\operatorname{d}\!t} &= \frac{\partial F_1(G_0,H_0)}{\partial G}g +\frac{\partial F_1(G_0,H_0)}{\partial H}h +J(t), \\
\frac{\operatorname{d}\!{h}}{\operatorname{d}\!t} &= \frac{\partial F_2(G_0,H_0)}{\partial G}g +\frac{\partial F_2(G_0,H_0)}{\partial H}h.
\end{align}

Zależności różniczkowe pomiędzy $g$ i $h$ mogą być linearyzowane w formie
\begin{align}
\frac{\operatorname{d}\!{g}}{\operatorname{d}\!t} &= -m_1g -m_2h +J(t), \tag{1} \\
\frac{\operatorname{d}\!{h}}{\operatorname{d}\!t} &= -m_3h +m_4g, \tag{2}
\end{align}
gdzie $m_1, m_2, m_3, m_4$ to pozytywne stałe, wszystkie wynoszące około $0.05$ [7].

W badaniu krzywej cukrowej interesuje nas tylko stężenie glukozy w krwi, więc pozbądźmy się zmiennej $h$. Zróżniczkujmy $\frac{\operatorname{d}\!{g}}{\operatorname{d}\!t}$ z $(1)$ względem $t$
\begin{equation}
\frac{\operatorname{d}^2g}{\operatorname{d}t^2}= -m_1\frac{\operatorname{d}\!{g}}{\operatorname{d}\!t} -m_2\frac{\operatorname{d}\!{h}}{\operatorname{d}\!t} +\frac{\operatorname{d}\!{J}}{\operatorname{d}\!t}.
\end{equation}
Podstawiając $\frac{\operatorname{d}\!{h}}{\operatorname{d}\!t}$ z $(2)$ oraz $m_2h$ z $(1)$ otrzymujemy
\begin{equation}
\frac{\operatorname{d}^2g}{\operatorname{d}t^2}= -(m_1+m_3)\frac{\operatorname{d}\!{g}}{\operatorname{d}\!t} -(m_1m_3 +m_2m_4)g +m_3J + \frac{\operatorname{d}\!{J}}{\operatorname{d}\!t},
\end{equation}
co możemy przepisać jako równanie
\begin{equation}
\frac{\operatorname{d}^2g}{\operatorname{d}t^2} + 2\alpha \frac{\operatorname{d}\!{g}}{\operatorname{d}\!t} + \omega_0^2g = S(t) \tag{3}
\end{equation}
ze stałymi $\alpha=\frac{1}{2}(m_1+m_3)\approx 0.05,~ \omega_0^2 = m_1m_3 + m_2m_4\approx 0.005$ i funkcją $S(t)=m_3J+\frac{\operatorname{d}\!{J}}{\operatorname{d}\!t}$.

Przypomnijmy, że $J(t)$ określa szybkość, z jaką zwiększa się stężenie glukozy we krwi, stąd równanie $(3)$ będzie niejednorodne tylko dla krótkiego interwału czasu po zażyciu glukozy.
Załóżmy, że dla czasu $t=0$ glukoza została całkowicie spożyta, wtedy dla $t\geqslant0, ~ g(t)$ spełnia
\begin{equation}
\frac{\operatorname{d}^2g}{\operatorname{d}t^2} + 2\alpha \frac{\operatorname{d}\!{g}}{\operatorname{d}\!t} + \omega_0^2g = 0. \tag{4}
\end{equation}

### **Rozwiązanie**
Równanie $(4)$ jest równaniem oscylacji [8]. Mamy wielomian charakterystyczny $r^2+2\alpha r + \omega_0^2$ z pierwiastkami $r=-\alpha \pm \sqrt{\alpha ^2 - \omega_0^2}\approx -0.05 \pm 0.05 i$. Zatem wszystkie rozwiązania $g(t)$ są postaci
\begin{equation}
g(t)=e^{-0.05t}[C_1\cos(0.05t) +C_2\sin(0.05t)].
\end{equation}

Stąd otrzymujemy przybliżony wzór $G(t)$
\begin{equation}
G(t) = G_0 + e^{-0.05t}[C_1\cos(0.05t) +C_2\sin(0.05t)].
\end{equation}
$G_0$ to wynik pomiaru glukozy we krwi na czczo pacjenta, natomiast niewiadome $C_1$ i $C_2$ wyznaczamy z n pomiarów $G_1,G_2,...,G_n$ dla $t_1,t_2,...,t_n$, tak, by błąd średniokwadratowy był zminimalizowany.

```python
import numpy as np
from scipy.optimize import minimize

# Dane wejściowe
t = np.array([0, 60, 120, 180, 240])
G_i = np.linspace(72, 99, 5) # wartości prawidłowe to zakres 72-99 mg/dl
G_0 = 86 # średnia z wartości prawidłowych

# Funkcja obliczająca błąd
def error(params):
    C1, C2 = params
    predicted_G = G_0 + np.exp(-0.05 * t) * (C1 * np.cos(0.05 * t) + C2 * np.sin(0.05 * t))
    error = np.mean(np.square(G_i - predicted_G))
    return error

# Początkowe wartości dla C1 i C2
initial_guess = [0,0]

# Minimalizacja błędu E
result = minimize(error, initial_guess)

C1_opt, C2_opt = result.x
print("Optymalne wartości:")
print(f"C1: {C1_opt}")
print(f"C2: {C2_opt}")
```

Weźmy $C_1 = -14$ oraz $C_2=1107$ ($C_2>0$, ponieważ wartości stężenia glukozy we krwi są nieujemne), ostatecznie otrzymujemy
\begin{equation}
\boxed{G(t) = G_0 + e^{-0.05t}[-14\cos(0.05t) + 1107\sin(0.05t)]}. \tag{5}
\end{equation}

# **5. ROZWIĄZANIA METODAMI NUMERYCZNYMI**

## **Metoda Eulera**

Metoda Eulera jest procedurą numeryczną pierwszego rzędu za pomocą której możemy rozwiąza równań różniczkowych zwyczajnych z zadaną wartością początkową Używamy małych stycznych na niewielkiej odległości, aby przybliżyć rozwiązanie problemu o wartości początkowej. Jest ona opisana wzorem rekurencyjnym[6]:
$$ y_{n} = y_{n-1}+hf(x_{n},y_{n}).$$
gdzie:
- $y_{n}$ to akualna wartość funkcji,
- $y_{n-1}$ to ostatnia wartość funkcji,
- $h$ to odstęp między krokami,
- $f(x_{n},y_{n})$ to wartość pochodnej funcji w punkcie $(x_{n},y_{n})$.

```python
# Parametry
alpha = 0.05
omega0_sq = 0.005
G0 = 92
t_end = 250
dt = 1 # Krok czasowy

# Inicjalizacja list
t_values = np.arange(0, t_end, dt)
g_euler = np.zeros_like(t_values, dtype=float)

g_euler[0] = G0


# Metoda Eulera
for i in range(1, len(t_values)):
    dgdt = -2 * alpha * g_euler[i-1] - omega0_sq * g_euler[i-1]
    g_euler[i] = G0 + g_euler[i-1] + dgdt * dt

# Rozwiązanie analityczne
def G_analytic(t, G0):
    return G0 + np.exp(-alpha * t) * (-14 * np.cos(np.sqrt(omega0_sq - alpha**2) * t) + 1107 * np.sin(np.sqrt(omega0_sq - alpha**2) * t))

g_analytic_values = G_analytic(t_values, G0)

# Wykres porównujący rozwiązania
plt.figure(figsize=(12, 6))
plt.plot(t_values, g_euler, label='Metoda Eulera', linestyle='--')
plt.plot(t_values, g_analytic_values, label='Rozwiązanie analityczne')
plt.xlabel('Czas (min)')
plt.ylabel('Poziom glukozy (mg/dl)')
plt.legend()
plt.title('Porównanie rozwiązania metodą Eulera i analitycznego')
plt.grid(True)
plt.show()

```

## **Metoda Rungego-Kutty**

Metoda Rungego-Kutty jest jedną z najpopularniejszych technik numerycznych do rozwiązywania równań różniczkowych. W przypadku metody 4. rzędu, wartości przybliżone są obliczane w czterech krokach, co daje wyższą dokładność w porównaniu do metod niższego rzędu.[9]
$$y_{i+1} = y_{i}+\frac{1}{6}(k_{1}+k_{2}+k_{3}+k_{4})h$$

gdzie:

$$k_{1}=f(x_{i},y_{i})$$
$$k_{2}= f(x_{i} + \frac{1}{2}h,y_{i} + \frac{1}{2}k_{1}h)$$
$$k_{3}=f(x_{i} +  \frac{1}{2}h,y_{i} + \frac{1}{2}k_{2}h)$$
$$k_{4}=f(x_{i}+h,y_{i} +k_{3}h)$$

```python
# Parametry
alpha = 0.05
omega_squared = 0.005
G0 = 92
t_end = 250
dt = 0.1

# Funkcje potrzebne do metody Rungego-Kutty
def dG_dt(t, G, dG):
    return dG

def d2G_dt2(t, G, dG):
    return -2 * alpha * dG - omega_squared * G

# Metoda Rungego-Kutty 4. rzędu
def runge_kutta_step(t, G, dG, dt):
    k1 = dt * dG_dt(t, G, dG)
    l1 = dt * d2G_dt2(t, G, dG)

    k2 = dt * dG_dt(t + 0.5 * dt, G + 0.5 * k1, dG + 0.5 * l1)
    l2 = dt * d2G_dt2(t + 0.5 * dt, G + 0.5 * k1, dG + 0.5 * l1)

    k3 = dt * dG_dt(t + 0.5 * dt, G + 0.5 * k2, dG + 0.5 * l2)
    l3 = dt * d2G_dt2(t + 0.5 * dt, G + 0.5 * k2, dG + 0.5 * l2)

    k4 = dt * dG_dt(t + dt, G + k3, dG + l3)
    l4 = dt * d2G_dt2(t + dt, G + k3, dG + l3)

    G_new = G + (k1 + 2*k2 + 2*k3 + k4) / 6
    dG_new =  dG + (l1 + 2*l2 + 2*l3 + l4) / 6

    return G_new, dG_new

# Symulacja
t_values = np.arange(0, t_end, dt)
G_values = np.zeros_like(t_values)
dG_values = np.zeros_like(t_values)
G_values[0] = G0
dG_values[0] = G0

for i in range(1, len(t_values)):
    G_values[i], dG_values[i] = runge_kutta_step(t_values[i-1], G_values[i-1], dG_values[i-1], dt)

# Rozwiązanie analityczne
def G_analytic(t, G0):
    return G0 + np.exp(-alpha * t) * (-14 * np.cos(np.sqrt(omega0_sq - alpha**2) * t) + 1107 * np.sin(np.sqrt(omega0_sq - alpha**2) * t))

G_analytic_values = G_analytic(t_values, G0)

# Wykres
plt.plot(t_values, G_values, label='Metoda Rungego-Kutty')
plt.plot(t_values, G_analytic_values, label='Rozwiązanie analityczne')
plt.xlabel('Czas (min)')
plt.ylabel('Poziom glukozy (mg/dl)')
plt.title('Porównanie rozwiązania metodą Rungego-Kutty i analitycznego')
plt.legend()
plt.grid(True)
plt.show()
# Dane dla wykresu funkcji G(t) z modelu
G_0 = 92
t_values = np.linspace(0, 250, 1000)
G_values = G_analytic(t_values, G_0)
```
